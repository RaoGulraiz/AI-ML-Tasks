```python
# --------------------------------------------
# Step 1: Import Required Libraries
# --------------------------------------------
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# To show plots inside the notebook
%matplotlib inline
```

```python
# --------------------------------------------
# Step 2: Load the Iris Dataset using seaborn
# --------------------------------------------
df = sns.load_dataset('iris')  # Or load CSV using pd.read_csv if downloaded

# Print first few rows
print("üîç First 5 Rows of the Iris Dataset:")
df.head()
```

```python
# --------------------------------------------
# Step 3: Basic Dataset Info
# --------------------------------------------

print("üìê Dataset Shape:", df.shape)
print("üìã Column Names:", df.columns.tolist())

# Dataset structure
print("\nüìò Dataset Info:")
print(df.info())

# Summary statistics
print("\nüìä Summary Statistics:")
df.describe()
```

```python
# --------------------------------------------
# Step 4: Scatter Plot - Sepal vs Petal Length
# --------------------------------------------
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title("üìà Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.show()
```

```python
# --------------------------------------------
# Step 5: Histograms - Feature Distributions
# --------------------------------------------
df.hist(figsize=(12, 8), edgecolor='black')
plt.suptitle("üìä Distribution of Iris Features", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# --------------------------------------------
# Step 6: Box Plots to Identify Outliers
# --------------------------------------------
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title("üì¶ Box Plot of Iris Features")
plt.grid(True)
plt.show()
```
